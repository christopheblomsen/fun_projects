### Estimer hvor mange sivilsasjoner
Du kan manuelt beregne hvor mange avanserte sivilisasjoner det er ved hjelp av Drake likningen
$$N = R^* \cdot f_p \cdot n_e \cdot f_l \cdot f_i \cdot f_c \cdot L $$

Hvor\
    $N$   = Antall sivilisasjoner i galaksen som vi kan oppdage den elektromagnetiske strålingen fra.\
    $R^*$ = Gjennomsnittelige raten som stjerner blir dannet. (Ny stjerne per år)\
    $f_p$ = Andell stjerner med planeter\
    $n_e$ = For stjerner med planeter, gjennomsnittelig andel planeter som har klima som er passende for liv. \
    $f_l$ = Andel planeter som har utviklet liv\
    $f_i$ = Andel planeter med liv som har intelligent og sivilert liv.\
    $f_c$ = Andel av sivilisasjonen som gir fra seg oppdagelige stråling om liv.\
    $L$   = Lengden av tid, i år, som sivilisasjonen sender ut oppdagelig stråling

Noen verdier som har vært opp gjennom

|Parameter|Drake 1961|Drake 2017|
|------|------|------|
|$R^*$|1|3|
|$f_p$|0.35|1|
|$n_e$|3|0.2|
|$f_l$|1|0.13|
|$f_i$|1|1|
|$f_c$|0.15|0.2|
|$L$|$50\times 10^6$|$1\times 10^9$|
|$N$|$7.9\times 10^6$|$15.6 \times 10^6$|

### Generer en sannsynlighet for oppdagelse formel
Hvis antall avanserte sivilasasjoner øker så øker sannsynligheten med at de oppdager hverandre. Det er kanskje veldig intuitivt, men hvordan finner vi den egentlige sannsynligheten da. Det fine med akkurat datamaskiner er at du kan bare hamre løs på problemet uten å sette opp noen super fancy algoritmer. Ene tilnærmingen her er da å lage en 3D model av melkeveien, så tilfeldig tildele sivilisasjoner gjennom. For å så beregne den Euklidske avstanden mellom dem, dere lærer om flere geometrier neste semester. Men at vi kan ha hundre av millioner av sivilisasjoner å analysere, så er denne måten litt datakraft dyr. 
Men siden vi håndterer ett stort sett med ukjente så er det ikke viktig å være super presis. Så vi kommer til å gjøre enda en tilnærming, fysikk er ikke lenger rene fasit svar, men hvilke tilnærminger du klarer å gjøre. Så kommer til å kutte bobblen opp i mindre biter sånn som sett på bildet under

Da kan vi finne volumet ved å buke, når $R$ er radiusen av galakse disken, og $r$ er radiusen til radio bobblen

\begin{align}
\text{disk volum} = \pi  R^2\cdot \text{disk høyde}\\
\text{radio bobble volum} = \frac{4}{3} \pi r^3\\
\text{skalert disk volum} = \frac{\text{disk volum}}{\text{radio bobble volum}} 
\end{align}

Den skalerte disk volumet er antall like volumer som "passer" inne i galaksen, du kan og bør se på disse boksene som numererte fra 1 til max antall bokser ofte skrevet som $N$. For å plasere en sivilisasjon så tar du veldig enkelt bare ett av disse boksnumerene tilfeldig, det finnes mange forskjellige måter å snakke om tilfeldighet, men vi kommer til å bruke `randint` metoden til Python. Hvis det blir flere som velger samme boks så betyr det så enkelt at det finnes flere sivilisasjoner i den samme boksen. Så antar vi at sivilisasjoner inne i samme bokser kan oppdage hverandre, dette er ikke helt sant, men har ikke tid til å gjøre dette dypere, men lang historie kort så kommer vi til å jobbe med veldig store tall, så summen kommer til å kanselere seg ut.

For å slippe å gjenta denne biten hver gang vi endrer hvor mange sivilisasjoner og eller radio bobblen, så kan vi lage en *polynomisk likning* som kan brukes til å genere alle fremtidige sannsynlighetsberegninger. Husk her at 1 tilsvarer da 100% når vi går videre.
