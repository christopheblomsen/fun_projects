## Fermi paradokset
Liten bakgrunnsinformasjon. Så har vi noe som heter *Drake equation* (https://en.wikipedia.org/wiki/Drake_equation). Som går ut på å finne hvor mange sivilisasjoner i galaksen som produserer elektromagnetisk stråling. I 2017 så ble den oppdatert med hensyn på nye planeter som ble funnet pga. Kepler sattelitten. For å kutte en lang historie kort så er det mindre enn 1 til 10 miliarder trilioner sannsynlighet for at vi er den første og eneste sivilisasjonen som kan klare dette. Så her kommer Fermi inn å sier noe ala "If they were out there, they'd be here." For i følge SETI instituttet så skulle man med selv en ganske oppnårlig rakett teknologi klare å utforske og eventuelt kolonisere melkeveien i løpet av 10 millioner år. Husk at vi ser på det astronomiske tidskallenderen her. Da er ikke det så stort det er faktisk bare 1/1000 dels av melkeveiens alder. Så målet er da at vi skal undersøke litt om mangelen på disse radio signalene med å se på sannsynligheten for at en sivilisasjon oppdager en annen.

### Estimer hvor mange sivilsasjoner
Du kan manuelt beregne hvor mange avanserte sivilisasjoner det er ved hjelp av Drake likningen
$$N = R^* \cdot f_p \cdot n_e \cdot f_l \cdot f_i \cdot f_c \cdot L $$

Hvor\
    $N$   = Antall sivilisasjoner i galaksen som vi kan oppdage den elektromagnetiske strålingen fra.\
    $R^*$ = Gjennomsnittelige raten som stjerner blir dannet. (Ny stjerne per år)\
    $f_p$ = Andell stjerner med planeter\
    $n_e$ = For stjerner med planeter, gjennomsnittelig andel planeter som har klima som er passende for liv. \
    $f_l$ = Andel planeter som har utviklet liv\
    $f_i$ = Andel planeter med liv som har intelligent og sivilert liv.\
    $f_c$ = Andel av sivilisasjonen som gir fra seg oppdagelige stråling om liv.\
    $L$   = Lengden av tid, i år, som sivilisasjonen sender ut oppdagelig stråling

Noen verdier som har vært opp gjennom

|Parameter|Drake 1961|Drake 2017|
|------|------|------|
|$R^*$|1|3|
|$f_p$|0.35|1|
|$n_e$|3|0.2|
|$f_l$|1|0.13|
|$f_i$|1|1|
|$f_c$|0.15|0.2|
|$L$|$50\times 10^6$|$1\times 10^9$|
|$N$|$7.9\times 10^6$|$15.6 \times 10^6$|

Da kan jeg informere om at vår egen radio bobble er på 225 Lyr, dette siden vi har en litt paranoid, ut ifra jordens historie, mot å møte på noe "folk" på egen jord. Så denne radiobølgen er da kun signaler som vi ikke kan stoppe, som tv, radio, etc. Men nå tenker du kanskje at 225 Lyr er veldig langt. Se på den figuren over melkeveien som du kan lage selv, og se at den bobblen er ganske liten. Videre så følger det den inverse andre kvadraters lov. Noe som betyr at hvis du dobbler avstanden så fire minsker du  energien til strålende. Så vi kunne bare oppdaget vår egen radiobobble på 16 Lyr.
